```python


import pandas as pd

LUAD_amp_df = pd.read_csv('LUAD_table_amp.conf_95.txt', sep='\t')
LUAD_amp_df['peak_id'] = 'chr' + LUAD_amp_df.chromosome.astype(str) + ':' + LUAD_amp_df.peak_start.astype(str) + '-' + LUAD_amp_df.peak_end.astype(str)
LUAD_amp_df.index = LUAD_amp_df['peak_id']
LUAD_del_df = pd.read_csv('LUAD_table_del.conf_95.txt', sep='\t')
LUAD_del_df['peak_id'] = 'chr' + LUAD_del_df.chromosome.astype(str) + ':' + LUAD_del_df.peak_start.astype(str) + '-' + LUAD_del_df.peak_end.astype(str)
LUAD_del_df.index = LUAD_del_df['peak_id']

LUSC_amp_df = pd.read_csv('LUSC_table_amp.conf_95.txt', sep='\t')
LUSC_amp_df['peak_id'] = 'chr' + LUSC_amp_df.chromosome.astype(str) + ':' + LUSC_amp_df.peak_start.astype(str) + '-' + LUSC_amp_df.peak_end.astype(str)
LUSC_amp_df.index = LUSC_amp_df['peak_id']
LUSC_del_df = pd.read_csv('LUSC_table_del.conf_95.txt', sep='\t')
LUSC_del_df['peak_id'] = 'chr' + LUSC_del_df.chromosome.astype(str) + ':' + LUSC_del_df.peak_start.astype(str) + '-' + LUSC_del_df.peak_end.astype(str)
LUSC_del_df.index = LUSC_del_df['peak_id']

SCLC_amp_df = pd.read_csv('SCLC_table_amp.conf_95.txt', sep='\t')
SCLC_amp_df['peak_id'] = 'chr' + SCLC_amp_df.chromosome.astype(str) + ':' + SCLC_amp_df.peak_start.astype(str) + '-' + SCLC_amp_df.peak_end.astype(str)
SCLC_amp_df.index = SCLC_amp_df['peak_id']
SCLC_del_df = pd.read_csv('SCLC_table_del.conf_95.txt', sep='\t')
SCLC_del_df['peak_id'] = 'chr' + SCLC_del_df.chromosome.astype(str) + ':' + SCLC_del_df.peak_start.astype(str) + '-' + SCLC_del_df.peak_end.astype(str)
SCLC_del_df.index = SCLC_del_df['peak_id']

LUAD_del_df
```

```python
tree_df = pd.read_excel('subclone_tree_edge.xlsx')
print(tree_df)

cnv_df = pd.read_excel('subclone_cnv.xlsx', index_col=0)
cnv_df.columns = [ '{}:{}-{}'.format(*x.split('.')) for x in cnv_df.columns]
cnv_df
```

```python
with open('tree_edge_focal_cnv.csv', 'w') as f:
    f.write('Type,Sample,Source,Target,Amp/Del,Peak,CN_Change,Gene\n')
    
    for i, row in tree_df.iterrows():
        #if i > 2: break
        cancer_type = row['Type']
        if cancer_type == 'LUAD':
            amp_df = LUAD_amp_df
            del_df = LUAD_del_df
        if cancer_type == 'LUSC':
            amp_df = LUSC_amp_df
            del_df = LUSC_del_df
        if cancer_type == 'SCLC':
            amp_df = SCLC_amp_df
            del_df = SCLC_del_df            
    
        source, target = row['Source'], row['Target']
    
        if source.endswith('_N'):
            source_cn = 2
        else:
            source_cn = cnv_df.loc[source]
        target_cn = cnv_df.loc[target]
        change = target_cn - source_cn
    
        amp_cn = change[change >= 1]
        del_cn = change[(change <= -0.5) & (target_cn < 2)]

        amp_bins = set(amp_cn.index) & set(amp_df.peak_id)
        del_bins = set(del_cn.index) & set(del_df.peak_id)

        for amp_bin in amp_bins:
            for gene in amp_df.loc[amp_bin].genes_in_peak.split(','):
                if gene:
                    f.write(','.join(map(str, [row['Type'], row['Sample'], source, target, 'Amp', amp_bin, amp_cn.loc[amp_bin], gene])) + '\n')
            
        for del_bin in del_bins:
            for gene in del_df.loc[del_bin].genes_in_peak.split(','):
                if gene:
                    f.write(','.join(map(str, [row['Type'], row['Sample'], source, target, 'Del', del_bin, del_cn.loc[del_bin], gene])) + '\n')
```
